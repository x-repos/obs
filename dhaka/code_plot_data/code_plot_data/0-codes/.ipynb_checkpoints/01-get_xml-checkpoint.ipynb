# How to get XML file?

```python
from obspy.clients.fdsn import Client

# from obspy import read, read_inventory
from obspy import UTCDateTime
# from glob import glob
```

The details about the seismic station could be find in the google. For example, 
the details of [AAK](https://www.fdsn.org/networks/detail/II/) could be find in google by typing seismic
station code AAK.

```python
client = Client("IRIS")
```

```python
stations=['AAK','BFO']
starttime=UTCDateTime("2023-02-06T00:00:00")
endtime=UTCDateTime("2023-02-07T23:59:59")
```

```python
for sta in stations:
    print(sta)
    inventory = client.get_stations(starttime=starttime, endtime=endtime,network="II", sta=str(sta),
                                    loc="*", channel="BHZ",level="response")
    inventory[0].plot_response(min_freq=1E-4)
    inventory.write("../staxml/"+str(sta)+".xml",format="STATIONXML")
    
# # pass band of instrument where all the frequencies signal amplify with same factor
# # not in pass band amplification depends on the frequency
# #lc hc
```
